```python
import numpy as np
import matplotlib.pyplot as plt

# ایجاد مقادیر x در بازه -5 تا 5 با 500 نقطه
x_values = np.linspace(-5, 5, 500)

# محاسبه مقادیر y
y_values1 = np.cos(1.2 * x_values)
y_values2 = np.exp(-x_values**2)

# ایجاد نمودار
plt.figure(figsize=(10, 6))
plt.title("Plot of cos(1.2x) and exp(-x^2)", fontsize=16)

# رسم نمودارها
plt.plot(x_values, y_values1, label='cos(1.2x)', color='blue', linestyle='-')
plt.plot(x_values, y_values2, label='exp(-x^2)', color='red', linestyle='--')

# افزودن برچسب‌ها و شبکه
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# نمایش نمودار
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ایجاد مقادیر x در بازه -5 تا 5 با 500 نقطه
x_values = np.linspace(-5, 5, 500)

# محاسبه مقادیر y = x^3
y_values = x_values**3

# ایجاد نمودار
plt.figure(figsize=(10, 6))
plt.title("Plot of x^3", fontsize=16)

# رسم خط x^3 به صورت خط‌چین سبز
plt.plot(x_values, y_values, color='green', linestyle='--', label='x^3')

# پر کردن زیر نمودار
plt.fill_between(x_values, y_values, color='cyan', alpha=0.3)

# افزودن برچسب‌ها و شبکه
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt
```
