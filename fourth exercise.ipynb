```python
import numpy as np
import matplotlib.pyplot as plt

def plot_normal_dist(mean=0, std=1, n=1000):
    """
           np.random.normal
    """
    data = np.random.normal(loc=mean, scale=std, size=n)

    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

    x = np.linspace(mean - 4*std, mean + 4*std, 1000)
    pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)
    plt.plot(x, pdf, 'r', linewidth=2, label='PDF')

    plt.title(f"Normal Distribution (n={n}, mean={mean}, std={std})")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

#    
plot_normal_dist(mean=0, std=1, n=10)
plot_normal_dist(mean=0, std=1, n=1000)
plot_normal_dist(mean=0, std=1, n=100000)
```
